```python
%matplotlib inline
import pandas as pd
import numpy as np
import math as math
import statsmodels.api as sm
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("data.xlsx")
data.head()
```

### Apenas a categoria Peso-Pesados ('Heavyweights')

```python
data.weight_class = data.weight_class.astype('category')
```

```python
data_heavy = data.loc[(data.weight_class=='Heavyweight'),:]
bool_to_number = {False: 0, True: 1}
string_to_number = {'Blue': 0, 'Red': 1}
data_heavy['title_bout'] = data_heavy['title_bout'].map(bool_to_number)
data_heavy['Winner'] = data_heavy['Winner'].map(string_to_number)
```

```python
data_heavy.head(2)
```

## Como fazer os Betas

```python
def preparo(X,Y):
    X_cp = sm.add_constant(X)
    model = sm.OLS(Y,X_cp)
    results = model.fit()
    return results
```

```python
Y = data_heavy["Winner"]
#X = data_heavy[['title_bout','R_Stance']]
X=data_heavy['title_bout']
np.asarray(X)
```

```python
result = preparo(X,Y)
result.summary()
```

```python
Beta=[]

Beta.append[result.params["const"]]
Beta.append[result.params['title_bout']]
Beta.append[result.params['R_Stance']]
```
